// pip install Faker 가짜 데이터 만들기

```python
import pandas as pd
from faker import Faker

fake = Faker('ko_KR') # 페이커 객체
```

```python
result01 = []
columns01 = ('아이디', '이름', '주소', '전화번호', '이메일')
myencoding = 'CP949' # 'CP949'  'UTF-8'  인코딩 문자열
```

```python
for idx in range(10):
    sublist = []
    sublist.append(fake.user_name())
    sublist.append(fake.name())
    sublist.append(fake.address())
    sublist.append(fake.phone_number())
    sublist.append(fake.email())
    result01.append(sublist)
```

<h6><font color="BlueViolet">데이터 프레임 내용 확인</font></h6>

```python
myframe01 = pd.DataFrame(result01, columns = columns01)
print('데이터 프레임 내용 확인')
myframe01
```

```python
filename = './../data/result01.csv'
myframe01.to_csv(filename , encoding=myencoding, mode='w', index=True)
```

<h2><font color="red">Pie 차트 그리기</font></h2>

<h4><font color="green">Pie 그래프(식비)</font></h4>

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
```

```python
chartdata01=payment['식비']
chartdata01.to_frame()
```

```python
mycolors = ['blue', '#6AFF00', 'yellow', '#FF003C', '#CCF0DD']
mylabel = chartdata01.index
chartdata01.plot(kind='pie', labels=mylabel, shadow=True, explode=(0, 0.05, 0, 0, 0),
        colors=mycolors, autopct='%1.2f%%', startangle=90, counterclock=False)
plt.grid(True)
plt.legend(loc=4)
plt.xlabel(chartdata01.index.name)
plt.title(chartdata01.name + ' 지급율')
```

<h4><font color="green">Pie 그래프(최종 식비)</font></h4>

```python
chartdata02=payment['최종 식비']
chartdata02.to_frame()
```

```python
mylabel = chartdata02.index
chartdata02.plot(kind='pie', labels=mylabel, shadow=False, explode=(0, 0.05, 0, 0, 0),
        colors=mycolors, autopct='%1.3f%%', startangle=0, counterclock=True)

plt.grid(True)
plt.legend(loc=4)
plt.xlabel(chartdata02.index.name)
plt.title(chartdata02.name + ' 지급율')
```
