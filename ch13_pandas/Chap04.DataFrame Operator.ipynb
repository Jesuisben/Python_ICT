<h6><font color="BlueViolet">조건식 필터링</font></h6>

```python
# 조회하고자 하는 부서 목록
dept_col = ['영업부', '전산부', '구매부']
# 조회하고자 하는 컬럼 목록
column_list = ['사원명', '출장지역', '부서', '성별', '교통비']

# 부서가 dept_col에 포함되어 있고, 남자이며, 교통비가 4000이상 7000이하인 데이터
# @ 기호는 치환 변수의 의미로 사용됩니다.
sql = "부서 in @dept_col and 성별 == '남자' and 4000 <= 교통비 <= 7000" # 검색 조건
resultSet = payment.query(sql)
resultSet[column_list]
```

<h6><font color="BlueViolet">빈도수가 높은 상위 부서 2개 고르기</font></h6>

```python
dept_frequency=payment.부서.value_counts()
print('부서별 빈도수 확인')
dept_frequency.to_frame().T
```

```python
rowsu=2 # 추출할 상위 개수
concern_data = dept_frequency.index[:rowsu].tolist()
concern_data
```

```python
# 빈도수 상위 2개 부서 멤버가 아닌 사원 중 남자인 사원은?
print('\n빈도 수가 많은 상위 ' + str(rowsu) + '개 : '+ str(concern_data))
filtering = payment.query("부서 not in @concern_data and 성별 == '남자'")
filtering
```
