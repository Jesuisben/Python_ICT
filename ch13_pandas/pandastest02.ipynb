```python
import pandas as pd
import numpy as np
```

```python
myindex02 = ["최유식", "윤선협", "유관철", "심형식", "연규희"]
mycolumns02 = ["중구", "용산구", "마포구", "노원구", "은평구"]
```

```python
mylist02 = list(10 * onedata for onedata in range(1, 26))
print(mylist02)
```

```python
myframe02 = pd.DataFrame(np.reshape(mylist02, (5, 5)),
                        index=myindex02,
                        columns=mycolumns02)
myframe02
```

```python
# DataFrame에서 찾기
# 열 색인 : 판다스의_DataFrame_객체.["열 색인 이름"]
# 행 색인 : 판다스의_DataFrame_객체.iloc[인덱스 숫자] / 판다스의_DataFrame_객체.loc[문자열]

# 인덱스 번호로 찾기
# 행으로 찾기  (열로 찾을때는 그냥 함수없이 판다스의_DataFrame_객체.["열 색인 이름"] 하면 됨)
# 문법 : 판다스의_DataFrame_객체.iloc[행, 열]
# 매개변수를 한개만 입력하면 행으로 인식함
# 열로 검색시 행에 대한 값도 적어줘야 함 ( : 기호는 모두를 뜻하는 기호)
myframe02.iloc[1] # 1번 행 색인
```

```python
myframe02.iloc[[1, 3]]
```

```python
myframe02.iloc[0::2]
```

```python
# 문자열로 찾기
myframe02.loc["연규희"]
```

```python
myframe02.loc[["최유식", "심형식"]]
```

```python
myframe02.loc["유관철":"연규희"]
```

```python
mytarget02 = np.random.choice(myframe02.index, 3)
mytarget02
```

```python
myframe02.loc[mytarget02]
```

```python
myframe02.loc[myframe02["중구"] > 100]
```

```python
# df는 매개변수 -> 여기서 매개변수는 myframe02 즉 DataFrame임
myframe02.loc[lambda df : df["마포구"] >= 130]
```

```python
print('"윤선협"과 "연규희"의 "용산구" 실적을 30으로 변경하기')
myframe02.loc[["유관철", "연규희"], ["용산구"]] = 30
myframe02
```

```python
# "최유식":"심형식" 은 이미  [] 이라서 []을 또 해줄 필요 없음
print("'최유식'부터 '심형식'까지 '노원구' 실적을 80으로 변경하기")
myframe02.loc["최유식":"심형식", ["노원구"]] = 80
myframe02
```

```python
# 매개변수는 행이 기본값이고 열이 옵션적인 매개변수여서
# 행으로 찾을때는 열에 아무것도 적지 않으면 기본값으로 모든 열이 나옴
print("'연규희'의 모든 실적을 50으로 변경하기")
myframe02.loc["연규희"] = 50
myframe02
```

```python
# 열로 찾을때도 행의 매개변수는 필수이므로 값을 넣어줘야하는데
# : 기호를 넣으면 모든 행 매개변수로 들어옴
print("모든 사람의 '은평구' 컬럼을 60으로 변경하기")
myframe02.loc[:, "은평구"] = 60
myframe02
```

```python
print("'용산구' 실적이 50이하인 데이터의 ['노원구', '은평구']를 모두 0으로 변경하기")
# myframe02["용산구"] <= 50 : "용산구" 열 색인에서 값이 50이하인 값들 (행들)
myframe02.loc[myframe02["용산구"] <= 50, ["노원구", "은평구"]] = 0
myframe02
```

```python
myindex31 = ['강유성', '안동철', '신호진']
mylist31 = [60, 30, 50]
myseries31 = pd.Series(data=mylist31, index=myindex31)
print('시리즈 출력 결과')
myseries31.to_frame()
```

```python
# 데이터 준비하기
```

```python
# myframe31 이름의 DataFrame 준비
myindex31 = ['강유성', '안동철', '이지성']
mycolumns31 = ['용산구', '마포구', '서대문구']
mylist31 = list(10 * onedata for onedata in range(1, 10))
print(mylist31)
```

```python
myframe31 = pd.DataFrame(np.reshape(np.array(mylist31), (3, 3)),
                     index=myindex31,
                     columns = mycolumns31 )
print('데이터프레임 출력 결과')
myframe31
```

```python
# DataFrame과 Series의 산수해보기
# axis = 0 : 행 이름과 맞춤 (Series의 index를 DataFrame의 행 index와 맞춘다)
# axis = 1 : 열 이름과 맞춤 (Series의 index를 DataFrame의 column 이름과 맞춘다)
myframe31.add(myseries31, axis=0)
```

```python
# myframe32 이름의 DataFrame 준비
myindex32 = ['강유성', '안동철', '정유미']
mycolumns32 = ['용산구', '마포구', '구로구']
mylist32 = list(5 * onedata for onedata in range(1, 10))
print(mylist32)
```

```python
myframe32 = pd.DataFrame(np.reshape(np.array(mylist32), (3, 3)),
                     index=myindex32,
                     columns = mycolumns32 )
print('데이터프레임 출력 결과')
myframe32
```

```python
# DataFrame끼리 산수해보기
# fill_value : 계산할때 한쪽에는 값이 있고 다른 한쪽에 값이 없으면 없는 쪽을 0으로 보고 계산
# 서로의 없는 값은 NaN으로 설정
myframe31.add(myframe32, fill_value=0)
```

```python
# 데이터 준비
myindex04 = ['강유성', '김영민', '김수진']
mycolumns04 = ['은평구', '용산구', '마포구']
mylist04 = [50, 60, 40, 80, 30, 50, 80, 40, 60]
myframe04 = pd.DataFrame(np.reshape(np.array(mylist04), \
                (3, 3)), index=myindex04, columns = mycolumns04 )
myframe04
```

```python
# 인덱스 (행) 추가하기 (reindexing : reindex() 함수 이용)
# "이지성" 추가
newindex04 = ["강유성", "이지성", "김영민", "김수진"]
result04 = pd.DataFrame(newindex04)
result04
```

```python
# 새로 추가한 행에 대한 열의 데이터는 없기에 당연히 NaN으로 표기됨
# fill_value로 기본값 넣기
result04 = myframe04.reindex(index = newindex04, fill_value=50)
result04
```

```python
# 새로 추가한 열에 대한 행의 데이터는 없기에 당연히 NaN으로 표기됨
# fill_value로 기본값 넣기
newcolumn04 = ["은평구", "용산구", "마포구", "강남구"]
result04 = myframe04.reindex(columns = newcolumn04, fill_value=40)
result04
```

```python
# 1. DataFrame에서 행 또는 열 삭제
# axis = 0 : 행 방향
# axis = 1 : 열 방향
# axis의 값에 직접 "index" 혹은 "columns"를 넣어도 됨 (기본값은 0(행 방향)임)
# 판다스_DataFrame객체.drop(["행 이름"], axis=0)
# 판다스_DataFrame객체.drop(["열 이름"], axis=1)
# 2. 여러 행 또는 열 삭제시 ["행 또는 열 이름", "행 또는 열 이름"]
# 이런식으로 콤마를 구분자로 열거하면 됨
result04.drop(["김영민"], axis=0)
```

```python
chartdata = myframe02.loc[:, ["중구", "마포구"]]
chartdata
```

```python
# matplotlib으로 그래프 만들기
import matplotlib.pyplot as plt
plt.rc("font", family="Malgun Gothic")
```

```python
chartdata.plot(title="사원별 구별 영업 실적", marker = "o", figsize=(10, 6))
plt.grid(True)
plt.ylim([0, 200])
plt.xlabel('영업 사원')
plt.ylabel('구별 실적')
```

```python
chartdata.plot(kind='bar', rot=0, title='사원별 구별 영업 실적 막대 그래프', \
               legend=True)

# 범례(legend)의 위치(location)는 알아서 최상('best')의 위치에 배치해주세요.
plt.legend(loc='best')
plt.xlabel('영업 사원')
plt.ylabel('구별 실적')
plt.ylim([0, 200])
```

```python
nrow, ncol = 1, 2 # 서브 플로팅 행과 열수

# fig는 그림(Figure) 객체, axs는 각 그림 영역을 저장하고 있는 배열
# subplots() : 반환타입이 2개
fig, axs = plt.subplots(nrow, ncol, figsize=(9, 3), sharey=True)
axs[0].bar(chartdata.index, chartdata['중구'])
axs[1].bar(chartdata.index, chartdata['마포구'])
fig.suptitle('구별 영업 실적')
```
