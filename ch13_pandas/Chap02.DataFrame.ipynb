<h4><font color="green">DataFrame에 쓰기</font></h4>

<h6><font color="BlueViolet">loc 속성 사용하기</font></h6>

```python
print('"윤선협"과 "연규희"의 "용산구" 실적을 30으로 변경하기')
myframe02.loc[['윤선협', '연규희'], ['용산구']] = 30
myframe02
```

```python
print('"최유식"부터 "심형식"까지 "노원구" 실적을 80으로 변경하시오.')
myframe02.loc['최유식':'심형식', ['노원구']] = 80
myframe02
```

```python
print('"연규희"의 모든 실적을 50으로 변경하기')
myframe02.loc[['연규희'], :] = 50
myframe02
```

```python
print('모든 사람의 "은평구" 컬럼을 60으로 변경하기')
myframe02.loc[:, ['은평구']] = 60
myframe02
```

```python
print('"용산구" 실적이 50이하인 데이터의 \'노원구\', \'은평구\']를 모두 0으로 변경하기')
myframe02.loc[myframe02['용산구'] <= 50, ['노원구', '은평구']] = 0
myframe02
```

<h2><font color="red">재색인과 색인 삭제 기능</font></h2>

<h4><font color="green">DataFrame 생성</font></h4>

<h6><font color="BlueViolet">회원별 지역별 영업 실적 데이터</font></h6>

```python
myindex04 = ['강유성', '김영민', '김수진']
mycolumns04 = ['은평구', '용산구', '마포구']
mylist04 = [50, 60, 40, 80, 30, 50, 80, 40, 60]
myframe04 = pd.DataFrame(np.reshape(np.array(mylist04), \
                (3, 3)), index=myindex04, columns = mycolumns04 )
myframe04
```

<h6><font color="BlueViolet">재색인(reindex())</font></h6>

```python
newindex04 = ['강유성', '이지성', '김영민', '김수진']
result04 = pd.DataFrame(newindex04)
result04
```

```python
result04 = myframe04.reindex(index = newindex04, fill_value = 50)
result04
```

```python
newcolumn04 = ['은평구', '용산구', '마포구', '강남구']
result04 = myframe04.reindex(columns = newcolumn04, fill_value = 40 )
result04
```

<h6><font color="BlueViolet">색인 삭제하기</font></h6>

```python
newframe04 = result04.drop(['김영민'], axis = 0)
newframe04
```

```python
newframe04 = result04.drop(['용산구', '강남구'], axis = 1)
newframe04
```

<h6><font color="BlueViolet">색인 지정(set_index())</font></h6>

```python
mydict04 = {
            '이름': ['김영민', '이지성', '강유성', '류지은', '신호진'], 
            '거주지': ['마포', '용산', '서대문', '마포', '용산'],
            '국어': [10 * idx for idx in range(1, 6)], 
            '영어': [10 * idx for idx in range(5, 0, -1)]
          }
myframe04 = pd.DataFrame(mydict04)
myframe04
```

```python
print('set_index는 컬럼으로 존재하던 항목을 색인으로 옮겨줍니다.')
newframe04 = myframe04.set_index(['이름'])
newframe04
```

```python
myseries04 = pd.Series([11 * idx for idx in range(1, 6)])
myseries04
```

```python
newframe04 = myframe04.set_index([myseries04])
newframe04
```

```python
print('drop=False 옵션 사용')
result04 = myframe04.set_index(['이름'], drop=False)
result04
```

<h6><font color="BlueViolet">색인 리셋(reset_index())</font></h6>

```python
print('reset_index() 메소드 사용하기')
newframe042 = newframe04.reset_index()
newframe042
```

```python
print('drop=True이면 색인에 존재하던 내용을 컬럼으로 옮기지 않습니다.')
newframe042 = newframe04.reset_index(drop=True)
newframe042
```

<h2><font color="red">DataFrame을 이용한 차트 그리기</font></h2>

<h4><font color="green">데이터 만들기</font></h4>

<h6><font color="BlueViolet">회원별 지역별 영업 실적 데이터</font></h6>

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
```

```python
chartdata=myframe02.loc[:, ['중구', '마포구']]
chartdata
```

<h4><font color="green">꺽은 선 그래프(다변량)</font></h4>

```python
chartdata.plot(title='사원별 구별 영업 실적 꺽은 선', figsize=(10,6), \
               legend=True, rot=0, marker='o')
plt.grid(True)
plt.ylim([0, 200])
plt.xlabel('영업 사원')
plt.ylabel('구별 실적')
```

<h4><font color="green">막대 그래프(요소 여러개)</font></h4>

```python
chartdata.plot(kind='bar', rot=0, title='사원별 구별 영업 실적 막대 그래프', \
               legend=True)

# 범례(legend)의 위치(location)는 알아서 최상('best')의 위치에 배치해주세요.
plt.legend(loc='best')
plt.xlabel('영업 사원')
plt.ylabel('구별 실적')
plt.ylim([0, 200])
```

<h4><font color="green">분할하여 그리기(subplotting)</font></h4>

```python
type(chartdata['중구'])
```

```python
nrow, ncol = 1, 2 # 서브 플로팅 행과 열수

# fig는 그림(Figure) 객체, axs는 각 그림 영역을 저장하고 있는 배열
fig, axs = plt.subplots(nrow, ncol, figsize=(9, 3), sharey=True)
axs[0].bar(chartdata.index, chartdata['중구'])
axs[1].bar(chartdata.index, chartdata['마포구'])
fig.suptitle('구별 영업 실적')
```
