<h3><font color="red">Series의 값 읽기/쓰기</font></h3>

<h4><font color="green">Series의 값 읽어 오기</font></h4>

<h6><font color="BlueViolet">시리즈 만들기</font></h6>

```python
myindex02 = ['프랑스', '독일', '중국', '미국', '일본', '영국', '호주']
mylist02 = [50, 30, 20, 60, 40, 80, 70]
myseries02 = pd.Series(data=mylist02, index=myindex02)
myseries02.to_frame()
```

<h6><font color="BlueViolet">문자열 label 사용</font></h6>

```python
print('색인의 이름으로 값 읽기(scalar 반환)')
print(myseries02['미국'])
```

```python
print('색인의 이름으로 값 읽기(Series로 반환)')
print(myseries02[['미국']])
```

```python
print('라벨 이름으로 슬라이싱')
print(myseries02['미국':'영국']) # 문자열 label :  from <= 범위 <= to
```

```python
print('여러 개의 색인 이름으로 데이터 읽기')
print(myseries02[['호주', '미국']])
```

<h6><font color="BlueViolet">숫자형 label 사용</font></h6>

```python
myseries02
```

```python
print('정수를 이용한 데이터 읽기')
print(myseries02.iloc[2])
```

```python
print('서로 떨어진 데이터 읽기')
print(myseries02[0:7:3])
```

```python
print('슬라이싱 사용하기')
print(myseries02[3:6]) # start <= 결과 < stop
```

<h4><font color="green">Series에 값 기록하기</font></h4>

<h6><font color="BlueViolet">숫자형 label 사용</font></h6>

```python
myseries02.iloc[2] = 22 # myseries02[2] = 22
myseries02[2:5] = 55
myseries02[['프랑스', '미국']] = 22
print('\'프랑스\'과 \'미국\'의 값 변경')
print(myseries02)
```

```python
myseries02[0::2] = 77
print('짝수번째 요소들만  77로 변경하기')
print(myseries02)
```

```python
print('50보다 작은 데이터만 추출하세요 : ')
result = myseries02[myseries02 < 50]
print(result)
```

<h2><font color="red">시리즈 연산하기</font></h2>

<h4><font color="green">Series 만들기</font></h4>

<h6><font color="BlueViolet">샘플용 시리즈 만들기</font></h6>

```python
myindex31 = ['일본', '영국', '호주', '독일']
mylist31 = [30, 40, 50, 60]

myindex32 = ['일본', '영국', '호주', '프랑스']
mylist32 = [20, 40, 60, 70]

myseries31 = pd.Series(data=mylist31, index=myindex31 )
myseries32 = pd.Series(data=mylist32, index=myindex32 )
```

```python
print('1번 시리즈')
print(myseries31)
```

```python
print('2번 시리즈')
print(myseries32)
```

<h4><font color="green">Series와 scalar의 산술 연산</font></h4>

<h6><font color="BlueViolet">산술 연산자와 공용 메소드의 사용</font></h6>

```python
su=10
print('시리즈에 ' + str(su) + ' 더하기')
print(myseries31 + su)
```

```python
print('add 메소드 사용하기')
print(myseries31.add(su))
```

```python
print('시리즈에 ' + str(su) + ' 빼기')
print(myseries31 - su)
```

```python
print('sub 메소드 사용하기')
print(myseries31.sub(su))
```

```python
print('시리즈에 ' + str(su) + ' 곱하기')
print(myseries31 * su)
```

```python
print('mul 메소드 사용하기')
print(myseries31.mul(su))
```

```python
print('시리즈에 ' + str(su) + ' 나누기')
print(myseries31 / su)
```

```python
print('floordiv 메소드 사용하기')
print(myseries31.floordiv(su)) # div()는 실수형, floordiv()는 정수형으로 반환해줍니다.
```

<h6><font color="BlueViolet">관계 연산자와 관련 메소드의 사용</font></h6>

```python
# 관계연산자 : > < >= <= ==(equal) !=(not equal)
# 실행 결과는 True/False 중 하나가 됩니다.
su=50
print('시리즈에 대한 관계(비교) 연산')
print(myseries31 >= su)
```

```python
myseries31[myseries31 >= 50]
```

```python
print('시리즈의 값이 50보다 큰지 파악해 봅니다.')
print(myseries31.gt(su))
```

```python
myseries31[myseries31.gt(50)]
```

```python
print('시리즈의 값이 50인가요?')
print(myseries31.eq(su))
```

<h4><font color="green">Series간의 산술 연산</font></h4>

<h6><font color="BlueViolet">산술 연산자 사용</font></h6>

```python
# NaN은 Not a Number의 줄인 표현입니다.
# 값이 없거나, 부정적인 개념의 의미로 사용이 됩니다.
# NaN이 포함되어 있으면 float64 타입으로 인식합니다.
print('한쪽에만 있는 데이터는 NaN으로 처리됨')
newseries = myseries31 + myseries32
print(newseries)
```

<h6><font color="BlueViolet">공용 메소드 사용하기</font></h6>

```python
print('두 시리즈  빼기(fill_value에 의하여 0으로 처리된 후 연산 수행됨)')
newseries = myseries31.sub(myseries32, fill_value = 0)
print(newseries)
```

```python
print('곱하기(mul)')
print('나누기(div)도 동일합니다.')
newseries = myseries31.mul(myseries32, fill_value = 0)
print(newseries)
```

<h2><font color="red">유일 값과 빈도 파악, 색인 필터링</font></h2>

<h4><font color="green">유일한 값 확인하기</font></h4>

<h6><font color="BlueViolet">unique() 메소드 사용</font></h6>

```python
mylist04 = ['미국', '영국', '영국','중국', '영국', '영국', '일본', '일본', '미국', '미국']
myseries04 = pd.Series(mylist04)
print('시리즈 정보 확인')
print(myseries04)
```

```python
myunique04 = myseries04.unique()
print('unique() 메소드 적용 결과')
print(myunique04)
```

<h4><font color="green">빈도 수 파악하기</font></h4>

<h6><font color="BlueViolet">value_counts() 메소드 사용</font></h6>

```python
print('항목별 빈도 개수')
print(myseries04.value_counts())
```

```python
print('항목별 빈도를 비율로 보여 주기')
print(myseries04.value_counts(normalize=True))
```

<h4><font color="green">색인 필터링</font></h4>

<h6><font color="BlueViolet">isin() 메소드 사용</font></h6>

```python
print('\'미국\'과 \'중국\' 데이터인가요? ')
mask04 = myseries04.isin(['미국', '중국'])
print(mask04)
```

```python
# 구해진 Boolean 형태의 mask04 정보를 이용하여 데이터를 필터링합니다.
print('조건에 맞는 데이터 찾기')
print(myseries04[mask04])
```
